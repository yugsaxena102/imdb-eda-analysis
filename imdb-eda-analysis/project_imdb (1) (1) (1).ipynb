# Q18. What is the average duration of movies by Certificate?


````python
import pandas as pd

# Load the dataset
df = pd.read_csv('movies.csv')

# Extract numeric minutes from duration column (e.g., '175 min' → 175)
df['duration_mins'] = df['duration'].str.extract(r'(\d+)').astype(float)

# Fill missing certificates
df['certificate'] = df['certificate'].fillna('Not Rated')

# Calculate average duration by certificate
avg_duration = (
    df.groupby('certificate')['duration_mins']
    .agg(avg_duration_min='mean', movie_count='count')
    .round(1)
    .sort_values('avg_duration_min', ascending=False)
    .reset_index()
)

print(avg_duration.to_string(index=False))
````

# Q19. Are ratings improving over the years?

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df = df.dropna(subset=['year', 'rating'])
df = df[(df['year'] >= 1950) & (df['year'] <= 2022)]

yearly = df.groupby('year')['rating'].mean().reset_index()
yearly.columns = ['year', 'avg_rating']

z = np.polyfit(yearly['year'], yearly['avg_rating'], 1)
````

````python
plt.figure(figsize=(10, 4))
plt.plot(yearly['year'], yearly['avg_rating'], color='steelblue', linewidth=1.5, label='Avg rating')
plt.plot(yearly['year'], np.poly1d(z)(yearly['year']), 'r--', linewidth=1.5, label=f'Trend ({z[0]:+.4f}/yr)')
plt.fill_between(yearly['year'], yearly['avg_rating'], alpha=0.1, color='steelblue')
plt.title('Are movie ratings improving over the years?')
plt.xlabel('Year')
plt.ylabel('Avg IMDb Rating')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('ratings_over_years.png', dpi=150)
plt.show()

print(f"Trend      : {z[0]:+.4f} per year")
print(f"1950–1970  : {yearly[yearly['year'] <= 1970]['avg_rating'].mean():.2f}")
print(f"2010–2022  : {yearly[yearly['year'] >= 2010]['avg_rating'].mean():.2f}")
````

# Q23. Which year had the highest total industry revenue?


````python
import pandas as pd

df = pd.read_csv('movies.csv')

# Clean gross_income
df['gross_income'] = pd.to_numeric(df['gross_income'].str.replace(',', ''), errors='coerce')
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df = df.dropna(subset=['year', 'gross_income'])

# Total revenue per year
revenue_by_year = df.groupby('year')['gross_income'].sum().sort_values(ascending=False)

print(revenue_by_year.head(10).apply(lambda x: f"${x:,.0f}").to_string())
````

# Q24. What is the best genre for a debut director? (Highest avg rating for 1st time directors)


````python
import pandas as pd

df = pd.read_csv('movies.csv')

# Find directors with only 1 movie (debut directors)
debut_directors = df.groupby('directors_name').filter(lambda x: len(x) == 1)

# Split genres and explode
debut_directors = debut_directors.copy()
debut_directors['genre'] = debut_directors['genre'].str.split(',')
debut_directors = debut_directors.explode('genre')
debut_directors['genre'] = debut_directors['genre'].str.strip()

# Avg rating by genre for debut directors
best_genre = (
    debut_directors.groupby('genre')['rating']
    .agg(avg_rating='mean', count='count')
    .query('count >= 10')
    .sort_values('avg_rating', ascending=False)
    .round(2)
    .reset_index()
)

print(best_genre.head(10).to_string(index=False))
````

# Q25. Do movies with more actors (ensemble cast) earn more?

````python
import pandas as pd
import numpy as np

df = pd.read_csv('movies.csv')

# ── clean shared columns ──────────────────────────────────────────
df['gross_income'] = pd.to_numeric(df['gross_income'].str.replace(',', ''), errors='coerce')
df['votes']        = pd.to_numeric(df['votes'].str.replace(',', ''), errors='coerce')
df['rating']       = pd.to_numeric(df['rating'], errors='coerce')


df['cast_size'] = df['stars_name'].str.split(',').str.len()
df['cast_group'] = pd.cut(df['cast_size'], bins=[0,1,2,3,4,10],
                          labels=['1','2','3','4','5+'])

q25 = (df.groupby('cast_group')['gross_income']
         .agg(avg_gross='mean', count='count')
         .dropna()
         .round(0))

print(q25.to_string())
````

# Q27. Analyse the "Sequel Effect" — Do movies with numbers in titles earn more?

# Q28. The Ultimate Business Decision: What movie should we fund?



````python
df['genre'] = df['genre'].str.split(',')
df_g = df.explode('genre')
df_g['genre'] = df_g['genre'].str.strip()

genre_stats = (df_g.groupby('genre')
                   .agg(avg_rating=('rating','mean'),
                        avg_gross=('gross_income','mean'),
                        hit_rate=('rating', lambda x: (x>7.5).mean()*100),
                        count=('rating','count'))
                   .query('count >= 50')
                   .round(2))

# Score = normalised rating + normalised gross + hit rate
for col in ['avg_rating','avg_gross','hit_rate']:
    genre_stats[col+'_norm'] = (genre_stats[col] - genre_stats[col].min()) / \
                                (genre_stats[col].max() - genre_stats[col].min())

genre_stats['score'] = (genre_stats['avg_rating_norm'] +
                        genre_stats['avg_gross_norm']  +
                        genre_stats['hit_rate_norm']) / 3

best = genre_stats['score'].idxmax()
top  = genre_stats.sort_values('score', ascending=False).head(5)[
       ['avg_rating','avg_gross','hit_rate','score']]

print(top.to_string())
print(f"\nFund a: {best} movie")
````
